# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

# Answer-3-To calculate the probability density function (PDF) of a normal distribution at a given point 'x' with a given mean(μ) and standard deviation (σ), we can use the formula for the PDF of the normal distribution:

```python
import math
```

```python
def normal_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -0.5 * ((x - mean) / std_dev) ** 2
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value
```

```python
mean = 0.0
std_dev = 1.0
point = 1.0
pdf_at_point = normal_pdf(point, mean, std_dev)
print(f'PDF at point {point} for mean={mean} and std_dev={std_dev}: {pdf_at_point}')
```

# Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

# Answer-6-To calculate the cumulative distribution function (CDF) of a Poisson distribution at a given point k with a given mean (λ), we can use the following formula for the Poisson CDF

```python
import math

def poisson_cdf(k, mean):
    cdf_value = 0.0
    for i in range(k + 1):
        cdf_value += (math.exp(-mean) * (mean ** i)) / math.factorial(i)
    return cdf_value

# Example usage
mean = 3.0
point = 5
cdf_at_point = poisson_cdf(point, mean)
print(f'CDF at point {point} for mean={mean}: {cdf_at_point}')
```

# Q7. How Binomial distribution different from Poisson distribution?

# Answer-7-The binomial distribution and the Poisson distribution are both probability distributions used to model discrete random variables, but they have different characteristics and are applied in different scenarios.

# Here are the key differences between the binomial distribution and the Poisson distribution:
# Number of Trials:Binomial Distribution: Models the number of successes in a fixed number of independent trials (experiments) with two possible outcomes (success or failure).Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence, without specifying a fixed number of trials.
# Outcome Probability:Binomial Distribution: Each trial has a constant probability of success (p) and failure (=1−q=1−p).Poisson Distribution: Represents events occurring at a constant average rate (λ), but the exact timing of events is random.
# Nature of Events:Binomial Distribution: Applicable when events are discrete and independent, with a clear binary outcome (success or failure).
# Poisson Distribution: Applicable when events are rare and occur independently over time or space, with no clear upper limit on the number of events that can occur.
# Parameters:Binomial Distribution: Defined by the number of trials (n) and the probability of success (p).Poisson Distribution: Defined by the average rate of occurrence (λ).
# Formulas:Binomial Distribution: Probability mass function (PMF) is given by (=)=()−P(X=k)=(kn)pkq n−k, where k is the number of successes.Poisson Distribution: Probability mass function (PMF) is given by (=)=−!P(X=k)= k!e −λ k, where k is the number of events.
# Shape:Binomial Distribution: Generally more symmetric and bell-shaped as the number of trials increases.Poisson Distribution: As the mean rate (λ) increases, it becomes more similar in shape to a normal distribution.

# Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

# Answer-10-In a normal distribution (also known as a Gaussian distribution), the least frequent data appears at the extremes or tails of the distribution, farthest away from the mean. The normal distribution is symmetric and bell-shaped, with the mean, median, and mode all located at the center of the distribution.

# The normal distribution is characterized by the following features:Mean (μ): The mean is the central point of the distribution and represents the highest point on the symmetrical bell-shaped curve. It is the average value around which the data is centered.

# Tails: The tails of the distribution represent the regions farthest away from the mean. These are the areas where the least frequent data points occur. The tails extend infinitely in both directions.

# Standard Deviation (σ): The standard deviation measures the spread or dispersion of the data points around the mean. In a normal distribution, about 68% of the data falls within one standard deviation from the mean, 95% within two standard deviations, and approximately 99.7% within three standard deviations.

# Skewness: A normal distribution has zero skewness, meaning it is perfectly symmetrical. If the distribution is skewed, the tail on the side of the longer or larger values will have fewer data points, making it the least frequent.
